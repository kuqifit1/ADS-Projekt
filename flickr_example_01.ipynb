### Flickr REST API (example 01): Get information from photos, get photos, plot photos

```python
from flickrapi import FlickrAPI
import os

# Working directory
os.chdir('U:/Lektionen/ADS_FS2021/Dokumente_KK/Week_02/Flickr')
print(os.getcwd())

# Keys for flickr API
FLICKR_PUBLIC = 'c6a2c45591d4973ff525042472446ca2'
FLICKR_SECRET = '202ffe6f387ce29b'

# Settings (extras = different resolution of photos)
flickr = FlickrAPI(FLICKR_PUBLIC, FLICKR_SECRET, format='parsed-json')
extras = 'url_sq,url_t,url_s,url_q,url_m,url_n,url_z,url_c,url_l,url_o'

# Get information about photos using the flickr API
cont = flickr.photos.search(text='orange butterflies', per_page=5, extras=extras)
photos = cont['photos']

# Object type
print(type(photos))

# Print information about photos (note than no photos are printed)
from pprint import pprint
pprint(photos)
```

```python
# Derive information about the first photo from the dictionary
import random

# Select one photo randomly
photo_01 = photos["photo"][random.randint(0,4)]
print(photo_01)
```

```python
# Photo url
url = photo_01["url_s"]
url
```

```python
# Download single photo
import urllib.request
urllib.request.urlretrieve(url,  'photo_01.jpg')
```

```python
# Plot the photo
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('photo_01.jpg')
plt.figure(figsize=(8,8))
plt.axis('off')
imgplot = plt.imshow(img)
```
